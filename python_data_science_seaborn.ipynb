```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())

g=sb.FacetGrid(df,col="day")
g.map(plt.hist,"tip")
plt.show()

g=sb.FacetGrid(df,col="sex")
g.map(plt.scatter,"total_bill","tip")
plt.show()

g=sb.barplot(x="sex",y="tip",data=df)
plt.show()

sb.regplot(x="total_bill",y="tip",data=df)
plt.show()

```

```python


sb.pairplot(df)
plt.show()
```

```python
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
sb.pairplot(df,hue='size')
plt.show()
```

```python
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
sb.pairplot(df,kind='reg')
plt.show()
```
